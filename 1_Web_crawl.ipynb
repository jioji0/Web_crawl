웹사이트에서 본문의 내용을 크롤링

```python
from langchain.document_loaders import WebBaseLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.chains.summarize import load_summarize_chain
from langchain.prompts import PromptTemplate

#백본에 관한 블로그
url = 'https://m.blog.naver.com/jin31/60006897907'
#웹 문서 크롤링
loader = WebBaseLoader(url)

#블로그 본문을 Chunk 단위로 쪼갬
text_splitter = CharacterTextSplitter(
    separator="\n\n",
    chunk_size=3000,      #쪼개는 글자수
    chunk_overlap=300,    #오버랩 글자수
    length_function=len,
    is_separator_regex=False,
)

#블로그 내용 크롤링 후 Chunk 단위로 분할
docs = WebBaseLoader(url).load_and_split(text_splitter)
docs
```


템플릿 정의

```python
from langchain import LLMChain
from langchain.prompts import PromptTemplate
from langchain.llms import HuggingFaceHub

# HuggingFace Repository ID
repo_id = 'openchat/openchat_3.5'

#각 Chunk 단위의 템플릿
template = '''다음의 내용을 한글로 요약해줘:

{text}
'''

#템플릿 생성
prompt = PromptTemplate(template=template, input_variables=['text'])

#llm 객체 생성
llm = HuggingFaceHub(
    repo_id=repo_id,
    model_kwargs={"temperature": 0.3,
                  "max_length": 128}
)
```

```python
!pip install tiktoken
```

```python
# 요약을 도와주는 load_summarize_chain
chain = load_summarize_chain(llm=llm,
                             map_prompt=prompt,
                             chain_type='map_reduce',
                             verbose=False)

# 실행결과
print(chain.run(docs))
```
